```python
kk=pd.merge(df,abc,on='Country Code',how='left')
kk.shape
```

```python
kk.dtypes
```

```python
country_names = kk.Country.value_counts().index
```

```python
country_values=kk.Country.value_counts().values
```

```python
#basics of data visuvalization - pie chart
plt.pie(country_values[:3], labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
#observation: Top 3 maximum sales of zomato is INDIA,USA,UK
```

```python
kk.columns
```

```python
ratings=kk.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
'''Observations
1. When rating is between 4.5 to 4.9  --> excellent
2. When rating is between 4.0 to 4.4  --> very good
3. When rating is between 3.5 to 3.9  --> good
4. when rating is between 2.5 to 3.4  --> average
5. when rating is between 1.8 to 2.4  --> poor
'''
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['pink','red','orange','yellow','green','green'])
```

```python
''' OBSERVATIONS:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4 '''
```

```python
#count plot
sns.countplot(x='Rating color',hue='Rating color',data=ratings,palette=['pink','red','orange','yellow','green','green'])
```
